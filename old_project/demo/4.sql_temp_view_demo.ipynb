#### Access dataframes using SQL
##### Objectives
1. Create temporary views on dataframes
2. Access the view from SQL cell
3. Access the view from Python cell

```python
%run "../includes/configuration"
```

```python
race_results_df = spark.read.parquet(f"{presentation_path}/race_results")
```

###### Create Temp View

```python
race_results_df.createOrReplaceTempView("v_race_results")
```

###### Access view with SQL

```python
%sql
select * from v_race_results where race_year = 2020
```

###### Access view with Python

```python
p_race_year = 2020
```

```python
race_results_2020 = spark.sql(f"select * from v_race_results where race_year = {p_race_year}")
```
