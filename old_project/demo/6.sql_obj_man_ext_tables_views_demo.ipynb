#### External Tables
##### Learning Objectives
1. Create external table using Python
1. Create external table using SQL
1. Effect of dropping an external table

```python
%sql
drop table if exists demo.race_results_ext_py;
```

```python
race_results_df.write.format("delta").option(f"presentation_path", "race_results_ext_py").saveAsTable("demo.race_results_ext_py")
```

```python
%sql
describe extended demo.race_results_ext_py
```

```python
%sql
drop table if exists demo.race_results_ext_sql
```

```python
%sql
create table demo.race_results_ext_sql
(race_year INT,
race_name STRING,
race_date TIMESTAMP,
circuit_location STRING,
driver_name STRING,
driver_number INT,
driver_nationality STRING,
team STRING,
grid INT,
fastest_lap INT,
race_time STRING,
points FLOAT,
position INT,
created_date TIMESTAMP
)
-- need ext location like s3
-- using delta
-- LOCATION "/Volumes/workspace/f1/processed/race_results_ext_sql"
```

```python
display(dbutils.fs.ls("/"))
```

```python
%sql
SHOW TABLES IN demo;
```

```python
%sql
INSERT INTO demo.race_results_ext_sql
SELECT * FROM demo.race_results_ext_py WHERE race_year = 2020;
```

```python
%sql
SELECT COUNT(1) FROM demo.race_results_ext_sql;
```

```python
%sql
DROP TABLE demo.race_results_ext_sql
```
