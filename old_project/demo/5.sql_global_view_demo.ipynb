###### Create Global View
####### not supported on serverless compute!

```python
race_results_df.createOrReplaceGlobalTempView("gv_race_results")
```

```python
%sql
SHOW TABLES IN global_temp;
```

```python
%sql
SELECT *
  FROM global_temp.gv_race_results;
```
