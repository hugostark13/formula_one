```python
%sql
select nationality, count(*) 
from drivers
group by nationality
order by nationality;
```

```python
%sql
select nationality, count(*) 
from drivers
group by nationality
having count(*) > 100
order by nationality;
```

```python
%sql
select nationality, name, dob, rank() over(partition by nationality order by dob desc) as age_rank from drivers order by nationality, age_rank limit 10;
```
