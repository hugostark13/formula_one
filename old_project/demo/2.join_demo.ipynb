```python
circuits_df = spark.read.parquet(f"{processed_path}/circuits") \
    .filter("circuit_id < 70") \
    .withColumnRenamed("name", "circuit_name")
```

```python
races_df = spark.read.parquet(f"{processed_path}/races").filter("race_year = 2019") \
    .withColumnRenamed("name", "race_name")
```

```python
display(circuits_df)
```

```python
display(races_df)
```

##### Inner join


```python
# Joins two dataframes and gives you records which satisfy 
# the contition, circuits_id in this case present in both dataframes
# and get the output from those
```

```python
races_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "inner") \
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round)
```

```python
display(races_circuits_df)
```

##### Outer Joins

```python
# Left dataframe is fixed, so you get every single recod from left dataframe
# and get the coresponding data from right dataframe
# if the value existing for the matching conditions
```

```python
# Left Outer Joins
races_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "left") \
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round).display()
```

```python
# Right dataframe is fixed, so you get every single recod from right dataframe
# and get the coresponding data from left dataframe
# if the value existing for the matching conditions
```

```python
# Right Outer Joins
races_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "right") \
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round).display()
```

```python
# Get data from both dataframes
# and if the condition in not satisfied
# either the left or right you get null values
```

```python
# Full Outer Joins
races_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "full") \
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round).display()
```

##### Semi Join

```python
# Only get records which satisfy the condition
# on both tables, but you get data from left dataframe
# and dont get any data from right dataframe
# tips: use inner join with select intead of this
```

```python
races_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "semi").display()
```

##### Anti Join

```python
# Gives the oposite of semi join
# every record from left dataframe which is not present in right dataframe
```

```python
races_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "anti").display()
```

##### Crross join

```python
# Takes every record from the left join to record on right
# and gives the result of two dataframes
```

```python
races_circuits_df = races_df.crossJoin(circuits_df)
```
