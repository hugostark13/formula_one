```sql
WITH driver_metrics AS
(
SELECT driver_name,
   SUM(race_starts) as race_starts,  
   SUM(number_of_wins) AS total_wins,
   SUM(number_of_podiums) AS total_podiums,
   SUM(CASE WHEN standing = 1 THEN 1 ELSE 0 END) AS total_championships
FROM formula_one.gold.v_driver_standing
GROUP BY driver_name
HAVING total_championships >=1
)
SELECT driver_name,
        race_starts,
        total_wins,
        total_podiums,
        total_championships,
        (total_championships * 100) + (total_wins * 10) + (total_podiums * 3) AS greathness_score
FROM driver_metrics
ORDER BY greathness_score DESC
```
